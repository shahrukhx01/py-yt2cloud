```python
from os import listdir
from os.path import isfile, join

```

```python
mypath = '/Users/shahrukh/Downloads/google_music/Music/'
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
import eyed3
import traceback
import os

try:
    for title in onlyfiles:
        if title != '.DS_Store':
            audiofile = eyed3.load('{}{}'.format(mypath,title))
            os.rename('{}{}'.format(mypath,title),'{}{} - {}.mp3'.format(mypath,audiofile.tag._getArtist(),audiofile.tag._getTitle())) 
            
    for f in listdir(mypath):
        print(f)
except Exception:
     traceback.print_exc()
```
